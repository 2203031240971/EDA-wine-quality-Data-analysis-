EDA DATA SET OF WINE (DATA ANALYSIS)


```python
import pandas as pd
import kagglehub



path = kagglehub.dataset_download("piyushagni5/white-wine-quality")

print("Path to dataset files:", path)
      


import os
csv_path = os.path.join(path, "winequality-white.csv")
df = pd.read_csv(csv_path)
df.head()



```

METHODS TO GET INFORMATION ON GIVEN DATA


```python
# Reload the CSV with the correct delimiter
df = pd.read_csv(csv_path, delimiter=';')

df.info()
df.describe()
df.shape
df.columns
df['quality'].unique()
##missing values in data set
df.isnull().sum()
#dulicate values in data set
df[df.duplicated()]
## removing duplicates values
df.drop_duplicates()
df.shape

df.corr()
```

DATA VISUALIZATION 

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(),annot=True)
## visulization
#it is an imbalanced data set
df.quality.value_counts().plot(kind='bar')
plt.xlabel('wine Quality')
plt.ylabel('Count')
plt.show()
```

```python
df.head()
for column in df.columns:
    sns.histplot(df[column],kde = True)

## see the perticular data 
sns.histplot(df['alcohol'])

## univariate,bivariate,multivariate analysis
sns.pairplot(df)


##  catogrical plot
sns.catplot(x='quality',y='alcohol',data=df,kind='box',color='green')
sns.scatterplot(x='quality', y='alcohol', data=df, color='blue')
```
